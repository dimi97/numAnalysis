###Κανόνας τραπεζίου

```python
def trapezoid(f,h,a,b):
  sum=(f(a)+f(b))*h/2
  x=a+h
  while x<b:
    sum+=f(x)*h
    x+=h 
  return sum
```

###Simpson

```python
def simpson(f,h,a,b):
  sum=(f(a)+f(b))*h/3
  x=a+h
  i=1
  while x<b:
    i+=1
    if i%2==0:
      sum+=4*f(x)*h/3
    else:
      sum+=2*f(x)*h/3
    x+=h

  return sum
```

### Romberg

```python
def romberg(f,h,a,b,k):
  I1=trapezoid(f,h,a,b)
  I2=trapezoid(f,k*h,a,b)

  A=(k**2*I1-I2)/(k**2-1)

  return A
```

### Error

```python
def error(real,res):
  return abs(real-res)
```

### Graph

```python
# importing the required module
import matplotlib.pyplot as plt

def graph(xValues,yValues):
  # x axis values
  x = xValues
  # corresponding y axis values
  y = yValues
    
  # plotting the points 
  plt.plot(x, y)
    
  # naming the x axis
  plt.xlabel('h Values')
  # naming the y axis
  plt.ylabel('Error')
    
  # giving a title to my graph
  plt.title('h-Error Graph')
    
  # function to show the plot
  plt.show()
```

### main

```python
import math
f=lambda x:10*math.exp(-50*abs(x))-(0.01/((x-0.5)**2+0.001))+5*math.sin(5*x)


real=-0.56682 #wolfram alpha


a=-1
b=1
k=2

n=2

errorsT=[]
errorsS=[]
errorsR=[]

hValues=[]

while n<=1000:
  h=abs(b-a)/n

  rt=trapezoid(f,h,a,b)
  rs=simpson(f,h,a,b)
  rr=romberg(f,h,a,b,k)

  errorT=error(real,rt)
  errorS=error(real,rs)
  errorR=error(real,rr)

  errorsT.append(errorT)
  errorsS.append(errorS)
  errorsR.append(errorR)

  hValues.append(h)

  #print("trapezoid",rt,"simpson",rs,"romberg",rr)
  #print("error trapezoid",round(100*errorT/abs(real),4),"% error simpson",round(100*errorS/abs(real),4),"% error romberg",round(100*errorR/abs(real),4),"%")

  n+=1
  
print(hValues)
print(errorsT)

graph(hValues,errorsT)
graph(hValues,errorsS)
graph(hValues,errorsR)

g=lambda x:x*math.exp(2*x)
```
